```python
import h5py
import numpy as np

def print_hdf5_structure(name, obj):
    if isinstance(obj, h5py.Dataset):
        print(f"Dataset: {name}")
        print(f"  Shape: {obj.shape}")
        print(f"  Type: {obj.dtype}")
        if len(obj.shape) > 0:
            print(f"  First few values: {obj[:5]}")  # Print first 5 values
    elif isinstance(obj, h5py.Group):
        print(f"Group: {name}")

def examine_hdf5(file_path):
    with h5py.File(file_path, 'r') as f:
        print(f"File: {file_path}")
        print("Structure:")
        f.visititems(print_hdf5_structure)

if __name__ == "__main__":
    file_path = "/global/cfs/cdirs/m3246/vmikuni/for_nishank/Aachen/train_ttbar.h5"  # Replace with the path to your file
    examine_hdf5(file_path)
```
