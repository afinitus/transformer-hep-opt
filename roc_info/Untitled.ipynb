```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Directory containing the ROC data files
roc_data_dir = '/pscratch/sd/n/nishank/humberto/transformer-hep/roc_info/aachen_preds'

# Get all NPZ files in the directory
roc_files = [f for f in os.listdir(roc_data_dir) if f.endswith('.npz')]

# Define a list of line styles and colors to match the legend style
linestyles = ['--', '-.', ':', '--', '-.', ':']
colors = ['blue', 'red', 'green', 'purple', 'brown', 'orange', 'black']

# Plot settings
plt.figure(figsize=(10, 8))
lw = 2

# Plot ROC curves for each model
for idx, roc_file in enumerate(roc_files):
    # Load the ROC data
    data = np.load(os.path.join(roc_data_dir, roc_file))
    
    # Extract predictions and labels
    predictions = data['predictions']
    labels = data['labels']
    
    # Compute ROC curve and ROC area
    fpr, tpr, _ = roc_curve(labels, predictions)
    roc_auc = auc(fpr, tpr)
    
    # Extract the N_train value from the filename (adjust this logic based on your actual filenames)
    # Assuming the filename contains something like 'N1000' or 'N1E3' for N_train values
    model_name = os.path.splitext(roc_file)[0]
    
    # Try to extract the N_train value from the filename
    try:
        # Assuming the N_train value is directly after 'N' in the filename (e.g., N1000_preds)
        n_train = int(''.join(filter(str.isdigit, model_name.split('_')[0])))
        n_train_sci = f'{n_train:.1E}'  # Scientific notation for N_train
    except:
        n_train_sci = 'N/A'  # Fallback if parsing fails
    
    # Plot the ROC curve (with TPR on x-axis and 1/FPR on y-axis)
    plt.plot(tpr, 1/fpr, lw=lw, linestyle='-', color=colors[idx % len(colors)], 
             label=rf'$N_{{train}} = {n_train_sci}, AUC={roc_auc:.5f}$')

# Plot settings
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xlim([0.0, 1.0])
plt.yscale('log')
plt.xlabel(r'$\text{TPR}$', fontsize=12)
plt.ylabel(r'$\frac{1}{\text{FPR}}$', fontsize=16)
plt.title('ROC Curve Comparison of Different Models')
plt.legend(loc="lower right", fontsize=10)

# Save the combined ROC plot
plt.savefig(os.path.join(roc_data_dir, 'combined_roc_curve.png'))

# Show the plot
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Paths to the .npy files
file_1 = '/global/cfs/cdirs/m3246/vmikuni/for_nishank/Aachen/npy/PET_opt_8_local_layer_scale_token_baseline_classifier.weights.npy'
file_2 = '/global/cfs/cdirs/m3246/vmikuni/for_nishank/Aachen/npy/PET_opt_8_local_layer_scale_token_fine_tune_classifier.weights.npy'

# Load the .npy files
data_1 = np.load(file_1, allow_pickle=True).item()
data_2 = np.load(file_2, allow_pickle=True).item()

# Extract ground truth labels and predicted probabilities
y_true_1 = data_1['y'][:, 1]  # Use the second column of the one-hot encoded labels
y_pred_1 = data_1['pred'][:, 1]  # Use the predicted probabilities for the positive class

y_true_2 = data_2['y'][:, 1]  # Use the second column of the one-hot encoded labels
y_pred_2 = data_2['pred'][:, 1]  # Use the predicted probabilities for the positive class

# Compute ROC curve and AUC for both files
fpr_1, tpr_1, _ = roc_curve(y_true_1, y_pred_1)
roc_auc_1 = auc(fpr_1, tpr_1)

fpr_2, tpr_2, _ = roc_curve(y_true_2, y_pred_2)
roc_auc_2 = auc(fpr_2, tpr_2)

# Plot settings
plt.figure(figsize=(10, 8))
lw = 2

# Plot ROC curves with TPR on the x-axis and 1/FPR on the y-axis
plt.plot(tpr_1, 1/fpr_1, lw=lw, linestyle='--', label=f'Baseline Classifier (AUC = {roc_auc_1:.5f})')
plt.plot(tpr_2, 1/fpr_2, lw=lw, linestyle='--', label=f'Fine-Tune Classifier (AUC = {roc_auc_2:.5f})')

# Plot settings
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xlim([0.0, 1.0])
plt.yscale('log')  # Log scale for the 1/FPR axis
plt.xlabel(r'$\text{TPR}$', fontsize=12)
plt.ylabel(r'$\frac{1}{\text{FPR}}$', fontsize=16)
plt.title('ROC Curve Comparison of Baseline and Fine-Tune Classifiers')
plt.legend(loc="lower right")

# Save the combined ROC plot
plt.savefig('combined_roc_curve.png')

# Show the plot
plt.show()
```
